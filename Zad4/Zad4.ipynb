```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('2021.csv', index_col="Country name")
df.head()
```

```python
plt.figure(figsize=(12, 5))
sns.lineplot(data=df["Ladder score"][:10])
plt.savefig('1.png')
```

```python
sns.histplot(data=df[:10],x="lowerwhisker",hue="Regional indicator")
plt.savefig('2.png')
```

```python
sns.kdeplot(data=df,x="Social support",hue="Regional indicator")
plt.savefig('3.png')
```

```python
sns.kdeplot(data=df,x="Ladder score",y="Social support", fill=True)
plt.savefig('4.png')
```

```python
sns.scatterplot(x=df["Generosity"], y=df["upperwhisker"])
plt.savefig('5.png')
```

```python
sns.scatterplot(x=df["Generosity"], y=df["lowerwhisker"], hue=df["Regional indicator"])
plt.savefig('6.png')
```
